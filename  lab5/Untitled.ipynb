```python
# basic model having only circle and rectangle and change to 2-3 colors
drawing = False # changes with mouse
mode = True # changes with buttons to arrange mode between circle and rectangle
ix,iy = -1,-1
color = (0,255,0)
# mouse callback function
def draw_circle(event,x,y,flags,param):
    global ix,iy,drawing,mode,color

    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix,iy = x,y

    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            if mode == True:
                cv2.rectangle(img,(ix,iy),(x,y),color,-1)
                #print(color)
            else:
                cv2.circle(img,(x,y),5,color,-1)
                #print(color)
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        if mode == True:
            cv2.rectangle(img,(ix,iy),(x,y),color,-1)
            #print(color)
        else:
            cv2.circle(img,(x,y),10,color,-1)
            #print(color)
            
img = np.zeros((512,512,3), np.uint8)
cv2.namedWindow('image')
cv2.setMouseCallback('image',draw_circle)
while(1):
    key = cv2.waitKey(1) 
    if key == ord('c'):
        mode = False
    elif key == ord('r'):
        mode = True
    elif key == ord('b'):
        color = (255,0,0)
    elif key == ord('q'):
        color = (0,0,255)
    elif key == ord('g'):
        color = (0,255,0)
    elif key == ord('w'):
        color = (255,255,255)
    print(color)
    cv2.imshow('image',img)
    
    
    if key & 0xFF == 27:
        break
    
cv2.destroyAllWindows()
```
